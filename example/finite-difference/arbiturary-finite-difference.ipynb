```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

# Preparation

### Construct coordinate

```python
x_arr = np.linspace(-1,1,11)
x_p_lim = (x_arr[0],x_arr[-1])
x_p_arr_input = np.array([-0.999, -0.3, 0.02, 0.41, 0.995, 1.0])
```

### Exclude out-of-range coordinates

```python
x_p_mask_arr = (x_p_lim[0] <= x_p_arr_input) & (x_p_arr_input < x_p_lim[1])
x_p_arr = x_p_arr_input[x_p_mask_arr]
```

### Prepare a test function : Gaussian wave packet

```python
from tdse.analytic import Gaussian1D, gradient_Gaussian1D, laplacian_Gaussian1D

k0 = 1.0
def get_psi_dpsi_ddpsi(x,t):
    _psi = Gaussian1D(x,t,k0)
    _dpsi = gradient_Gaussian1D(x,t,k0)
    _ddpsi = laplacian_Gaussian1D(x,t,k0)
    return (_psi, _dpsi, _ddpsi)

ana_list = get_psi_dpsi_ddpsi(x_arr, t=0.2)
sf_arr = ana_list[0]
```

### Configure the finite-difference method

```python
deriv_orders = [0,1,2]  # the orders of target derivatives
N_s = 4  # the number of stencils
```

# Evaluate by using arbiturary-stencils finite-difference method

```python
from tdse.finite_difference import eval_deriv_on_equidistanced_grid
result_list = eval_deriv_on_equidistanced_grid(x_p_arr, x_arr, sf_arr, deriv_orders, N_s)
```

# Plot the result

```python
rcParams['font.size'] = 15

fig, ax = plt.subplots(figsize=(10,4))

for _f, _f_p, _order in zip(ana_list, result_list, deriv_orders):
    line, = ax.plot(x_arr, _f.real, '.-', label="$f^{{({:d})}}(x)$".format(_order))
    ax.plot(x_p_arr, _f_p.real, 'x', color=line.get_color())

ax.set_xlabel("x"), ax.set_ylabel("function value")
ax.legend()
fig.tight_layout()
    
fig.savefig("fd-gaussian.png")
```
